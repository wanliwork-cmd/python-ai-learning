```python
import numpy as np
import torch as th
```

```python
#使用标量创建张量
t1 = th.tensor(data=100)
t1.dtype  #torch.int64  # 创建的张量类型为int64
```

```python
# 使用numpy中的ndarray数组来创建张量
arr1 = np.random.randn(3,3)  # 创建一个3x3的随机numpy数组，均值为0，标准差为1
t2 = th.tensor(arr1)
t2.dtype  #torch.float64  #类型为float64
```

```python
#使用容器来创建张量
t3 = th.tensor(data=[1,2,3,4,5])
t3.dtype  #torch.int64  # 创建的张量类型为int64
```

```python
print(t1)
print("======================")
print(t2)
print("======================")
print(t3)
```
